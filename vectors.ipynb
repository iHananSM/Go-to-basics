# Vectors 

notes:
* Data represented as vector
* **Vector space:** set of vector where you can perform operations like **pairwise addition**, and **scalar multiplication** that resulted vector on the same space.
* **Elements:** vectors (things can be added and scaled)
* Vectors could be:
    -   Geometric
    -   Polynomial: Vector space: all of (measurable) functions = function space
    -   Signals
    -   Elements of $R^n$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#R^2 Vectors
v1 = np.array([2, 4])
v2 = np.array([1, -3])

plt.figure()
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label="Vector A")
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', label="Vector B")

plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.axhline(0, color='grey', lw=0.5)
plt.axvline(0, color='grey', lw=0.5)
plt.legend()
plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Vector A and Vector B")
plt.show()
```

When visualizing vectors, they are typically represented as arrows that start at the origin (0,0) and point to the coordinates defined by the vector's values.
This helps in understanding the vector’s direction and magnitude
- **Direction:** indicates the vector’s orientation in space relative to the origin.
              e.g. A vector $\vec{v_1}=[2, 4]$ points from the origin $(0,0)$ to the point $(2,4)$.
- **Magnitude:** The length of the arrow (vector), for a vector $\vec{v}=[x, y]$, the magnitude $||\vec{v_2}|| = \sqrt{x^2 + y^2}$

### Vectors Operations

```python
# Vector addition
vector_sum = v1 + v2

# Plot vectors and their sum
plt.figure()
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label="Vector A")
plt.quiver(0, 0, v1[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', label="Vector B")
plt.quiver(0, 0, vector_sum[0], vector_sum[1], angles='xy', scale_units='xy', scale=1, color='g', label="A + B")

plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.axhline(0, color='grey', lw=0.5)
plt.axvline(0, color='grey', lw=0.5)
plt.legend()
plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Vector Addition (A + B)")
plt.show()
```

```python
scalar_small = 0.5
scalar_large = 2
vector_scaled_small = scalar_small * v1
vector_scaled_large = scalar_large * v1

# Plot original and scaled vectors
plt.figure()
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label="Vector A")
plt.quiver(0, 0, vector_scaled_small[0], vector_scaled_small[1], angles='xy', scale_units='xy', scale=1, color='b', label="0.5 * A")
plt.quiver(0, 0, vector_scaled_large[0], vector_scaled_large[1], angles='xy', scale_units='xy', scale=1, color='g', label="2 * A")

plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.axhline(0, color='grey', lw=0.5)
plt.axvline(0, color='grey', lw=0.5)
plt.legend()
plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scalar Multiplication of Vector A")
plt.show()
```
