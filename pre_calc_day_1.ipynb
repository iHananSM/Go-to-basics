### Basics on Python

```python
import matplotlib.pyplot as plt
from matplotlib.path import Path
import matplotlib.patches as patches
```

```python
verts = []
steps = []
```

```python
# start point

verts.append((0., 0.))
steps.append(Path.MOVETO)
```

```python
# first point

verts.append((2., 2.))
steps.append(Path.LINETO)
```

```python
verts.append((2., -2.))
steps.append(Path.LINETO)
```

```python
verts.append((0., 0.))
steps.append(Path.CLOSEPOLY)
```

```python
fig, ax = plt.subplots()
plt.axis([-5, 5, -5, 5])
path = Path(verts, steps)
patch = patches.PathPatch(path, facecolor='orange', lw=2)
ax.add_patch(patch)
plt.show()
```

```python
verts = [(0., 0.),
         (-1., 1.),
         (0., 2.),
         (2.,2.),
         (3., 1.),
         (2., 0.),
         (0., 0.)]

steps = [Path.MOVETO,
         Path.LINETO,
         Path.LINETO,
         Path.LINETO,
         Path.LINETO,
         Path.LINETO,
         Path.CLOSEPOLY]

fig, ax = plt.subplots()
plt.axis([-5, 5, -5, 5])
path = Path(verts, steps)
patch = patches.PathPatch(path, facecolor='blue', lw=2)
ax.add_patch(patch)
plt.show()
```

### Trigonometry

In tringle, for each Angle $A$, we have:
- Opposite Side (shorter)
- Adjecent Side
- Hypotenous Side (longest)

And the relationships (ratio) are:
- $Sine(A) = OPP/HYP$
- $Cosine(A) = ADJ/HYP$
- $Tangent = OPP/ADJ$

```python
import matplotlib.pyplot as plt
from IPython import display
fig, ax = plt.subplots()
plt.axis([-2, 12, -2, 8])
ax.text(0, 1, 'A')
ax.plot([1], [1], 'ro')
ax.plot([1,7], [1,1], 'b')
ax.plot([7,7], [1, 4], 'b')
ax.plot([1, 7], [1, 4], 'b')

display.display(plt.gcf())
display.clear_output(wait=True)

ax.text(2, 0, 'Adjecent Side')
display.display(plt.gcf())
display.clear_output(wait=True)

ax.text(8, 2, 'Opposite Side')
display.display(plt.gcf())
display.clear_output(wait=True)

ax.text(1, 3, 'Hypotenous Side')
display.display(plt.gcf())
display.clear_output(wait=True)
plt.show()
```

```python
import math

#angle in degree
a = 30

sine = math.sin(math.radians(a))
cosine = math.cos(math.radians(a))
tangent = math.tan(math.radians(a))

print(f'Sine: {round(sine, 4)}')
print(f'Cosine: {round(cosine, 4)}')
print(f'Tangent: {round(tangent, 4)}')
```

### Complmentory Angel

```python
#90
b = 90 - a

sine_b = math.sin(math.radians(b))
cosine_b = math.cos(math.radians(b))
tangent_b = math.tan(math.radians(b))

print(f'Sine: {round(sine_b, 4)}')
print(f'Cosine: {round(cosine_b, 4)}')
print(f'Tangent: {round(tangent_b, 4)}')
```
